```python
import pandas as pd

df = pd.read_csv('C:\\Users\\Abdullah\\Desktop\\Code\\Sentiment Based Stock\\Sentiment Based Stocks\\results\\final_dataset.csv')
```

```python
df['future_close_3h'] = df['close'].shift(-3)
df['return_3h']       = (df['future_close_3h'] - df['close']) / df['close']
df['label_3h']        = (df['return_3h'] > 0).astype(int)
```

```python
df['future_close_6h'] = df['close'].shift(-6)
df['return_6h']       = (df['future_close_6h'] - df['close']) / df['close']
df['label_6h']        = (df['return_6h'] > 0).astype(int)
```

```python
df.dropna(inplace=True)
```

```python
df.to_csv('C:\\Users\\Abdullah\\Desktop\\Code\\Sentiment Based Stock\\Sentiment Based Stocks\\results\\final_dataset_labeled.csv', index=False)
print("Shape:", df.shape)
```
