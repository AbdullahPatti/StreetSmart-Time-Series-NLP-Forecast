```python
import yfinance as yf
import pandas as pd
from ta.momentum import RSIIndicator
from ta.trend import SMAIndicator
```

```python
btc = yf.download("BTC-USD", interval="1h", period="2y")
```

```python
btc.columns = [c[0].lower() if isinstance(c, tuple) else c.lower() for c in btc.columns]
btc.index = pd.to_datetime(btc.index).tz_localize(None)
btc.reset_index(inplace=True)
btc.rename(columns={'datetime': 'hour'}, inplace=True)
```

```python
btc['returns']    = btc['close'].pct_change()
btc['volatility'] = btc['returns'].rolling(24).std()
btc['rsi']        = RSIIndicator(btc['close'], window=14).rsi()
btc['ma_7']       = SMAIndicator(btc['close'], window=7).sma_indicator()
btc['ma_30']      = SMAIndicator(btc['close'], window=30).sma_indicator()
btc['regime']     = (btc['volatility'] > btc['volatility'].median()).astype(int)
btc.dropna(inplace=True)
```

```python
btc.to_csv('btc_hourly_features.csv', index=False)
print("BTC shape:", btc.shape)
```
